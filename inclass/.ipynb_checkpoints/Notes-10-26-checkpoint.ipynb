```python
%matplotlib inline
import numpy as np
import pylab
import urllib

# let's fetch the data
urllib.request.urlretrieve('https://goo.gl/mjDtMB', 'hourly.pkl')

import pickle
hourly = pickle.load(open('hourly.pkl', 'rb'))
hourly[:10]
```

Now we can ask questions of these data. 

```python
counts = hourly[:,0]
months = hourly[:,1]
days = hourly[:,2]
hours = hourly[:,3]
dow = hourly[:,4]

pylab.plot(counts)
```

```python
dcounts = np.reshape(counts, (-1, 24))
dcounts[:2]
```

```python
pylab.plot(np.sum(dcounts, axis=1))
```

We wrote a function to get the total number of books read on a given day of the week.

```python
def BooksOnDOW(data, dow):
    return np.sum(data[data[:,4] == dow, 0])
```

Books read on Monday.

```python
BooksOnDOW(hourly, 0)
```

```python
BooksOnDOW(hourly, 2)
```
