Now I just need to add some code to track the longest run.

I see people use this approach so I will too. I don't particularly like it.

```python
def F1(A, N):
    run = 0
    runs = []
    for i in range(len(A)):
        if A[i,0] < N:
            run = run + 1
            runs.append(run)
            #print(i, run)
        else:
            run = 0
    return max(runs)
F1(hourly, 10)
```

I think this one is better. No list to keep track of. 

```python
def F1(A, N):
    run = 0
    maxrun = 0
    for i in range(len(A)):
        if A[i,0] < N:
            run = run + 1
            if run > maxrun:
                maxrun = run
            #print(i, run)
        else:
            run = 0
    return maxrun
F1(hourly, 10)
```

```python
def countVals(A):
    R = np.zeros(np.max(A)+1, dtype=int)
    for i in range(len(A)):
        R[A[i]] = R[A[i]] + 1
    return R
countVals(B)
```

Longest run of increasing numbers of books read.

```python
def F2(A):
    run = 0
    maxrun = 0
    for i in range(1, len(A)):
        if A[i,0] > A[i-1,0]:
            run = run + 1
            if run > maxrun:
                maxrun = run
            #print(i, run)
        else:
            run = 0
    return maxrun
F2(hourly)   
```

Return the longest sequence

```python
def F2(A):
    run = 0
    maxrun = 0
    maxindex = 0
    for i in range(1, len(A)):
        if A[i,0] > A[i-1,0]:
            run = run + 1
            if run > maxrun:
                maxrun = run
                maxindex = i
            #print(i, run)
        else:
            run = 0
    print(maxrun, maxindex)
    return A[maxindex-maxrun:maxindex+1]
F2(hourly)   
```
