## 1. How many days were _beautiful_ at each station?

Where _beautiful_ is defined to be rain free with temperature no less than 60 degrees and dew point no greater than 60 degrees from 7am through 7pm inclusive.

Write a function `F1` that returns the number of beautiful days at each station given `temp`, `dewp`, and `prcp` as arguments.

```python
def F1(temp,dewp,prcp):
    beauttemp = temp[:,:,7:20] >= 60
    beautdewp = dewp[:,:,7:20] <= 60
    beautprcp = prcp[:,:,7:20] == 0
    beautday = np.all(beauttemp & beautdewp & beautprcp, axis=2)
    return np.sum(beautday, axis=1)

def F1A(temp,dewp,prcp):
    beautifuldays = np.zeros(temp.shape[0], dtype=int)
    for i in range(temp.shape[0]):
        for j in range(temp.shape[1]):
            if np.all(temp[i,j,7:20] >=60) and np.all(dewp[i,j,7:20]<=60) and np.all(prcp[i,j,7:20] ==0):
                beautifuldays[i]+=1
    return beautifuldays


# let's see what your function returns
print('F1 returns', F1(temp, dewp, prcp))

check('Q1', F1, temp, dewp, prcp, points=10)


```

## 2. What was the maximum temperature range in a single day for each station?

Write a function F2 that returns the maximum temperature range for each station given the `temp` array. That is, you want to find the difference between the max and min temperatures for each day and return the max of those differences.

```python
def F2(temparray):
    temprange = np.zeros(temparray.shape[0])
    localmax = 0
    localmin = 1000
    biggestrange = 0
    currentrange = 0
    for i in range(temparray.shape[0]):
        for j in range(temparray.shape[1]):
            for k in range(temparray.shape[2]):
                if temparray[i,j,k]> localmax:
                    localmax = temparray[i,j,k]
                if temparray[i,j,k]< localmin:
                    localmin = temparray[i,j,k]
            currentrange = localmax- localmin
            if currentrange > biggestrange:
                biggestrange = currentrange
            localmax = 0
            localmin = 1000
        temprange[i]=biggestrange
        biggestrange=0
        currentrange=0
    return temprange        

def F2A(temp):
    maxes = np.max(temp, axis=2)
    mins = np.min(temp, axis=2)
    ranges = maxes-mins
    return np.max(ranges, axis=1)

def F2B(temp):
    ranges = np.zeros(temp.shape[0])
    for i in range(temp.shape[0]):
        currentmax = 0
        for j in range(temp.shape[1]):
            rangeday= np.max(temp[i,j,:])-np.min(temp[i,j,:])
            if rangeday > currentmax:
                currentmax = rangeday
        ranges[i]=currentmax
    return ranges
            
    
    
    
# let's see what your function returns
print('F2 returns', F2(temp))

check('Q2', F2, temp, points=10)
```

```python
#32: Is the second half of the year statistcally warmer than the first half of the year? Aka calculate the p value with N number of tests
def F32(temp, N):
    half = temp.shape[1]//2
    firsthalf = temp[:,0:half,:]
    secondhalf = temp[:,half:,:]
    firstmean = np.mean(firsthalf)
    secondmean = np.mean(secondhalf)
    difference = secondmean-firstmean
    counter = 0
    for i in range(N):
        temp2= np.random.permutation(temp)
        firsthalf2 = temp2[:,0:half,:]
        secondhalf2 = temp2[:,half:,:]
        firstmean2 = np.mean(firsthalf2)
        secondmean2 = np.mean(secondhalf2)
        difference2 = secondmean2-firstmean2
        if difference2 > difference:
            counter += 1
    return counter/N
F32(temp, 10000)
```

```python
#41: Write a function that takes an average daily dewpoint, temperature, prcp, starting day, and ending day and gives the number of days that have less than those values
def F41(temp, dewp, prcp, maxt, maxd, maxp, sday, eday):
    tempavg = np.mean(temp, axis =2)
    dewpavg = np.mean(dewp, axis =2)
    prcpavg = np.mean(prcp, axis =2)
    tdays = tempavg[:,sday:eday+1] <= maxt
    ddays = dewpavg[:,sday:eday+1] <= maxd
    pdays = prcpavg[:,sday:eday+1] <= maxp
    alldays = (tdays & ddays & pdays)
    return np.sum(alldays, axis=1)
F41(temp, dewp, prcp, 70, 60, 5, 3, 302)
```
