```python
%matplotlib inline
import numpy as np
import pylab

N = 20  # the number of points

np.random.seed(1) # so we get the same results

# create random points over the range -1 to 1
A = np.random.sample((N,2)) * 2 - 1

print(A)

pylab.plot(A[:,0], A[:,1], 'o')
pylab.axis('equal')
```
