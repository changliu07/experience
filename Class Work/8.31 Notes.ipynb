```python
%matplotlib inline
import numpy as np
import pylab as pl
```

```python
x = np.linspace(0, 1, 101)  # step from 0 to 1 in 101 steps
y = np.sin(4 * np.pi * x) * np.exp(-5 * x) # a damped sine

pl.plot(x, y)
pl.grid(True)
```
